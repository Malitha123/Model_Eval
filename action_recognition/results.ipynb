```python
import pandas as pd
import os
```

```python
direct = {'direct':"/nfs/users/ext_malitha.gunawardhana/SEVERE-BENCHMARK/action_recognition/jobs/direct_evaluation_results",
    'linear':'/nfs/users/ext_malitha.gunawardhana/SEVERE-BENCHMARK/action_recognition/jobs/linear_fine_tune_results',
'full_finetune':'/nfs/users/ext_malitha.gunawardhana/SEVERE-BENCHMARK/action_recognition/jobs/full_fine_tune_evaluation_results'}

eval, Model, Date, Time, Video_at_1, Video_at_MeanTop1, Video_at_5, Video_at_MeanTop5 = [], [], [], [], [], [], [], []
for main in direct.keys():
    path = os.listdir(direct[main])
    for i in range(len(path)):
        file = direct[main]+ '/'+ path[i]
        model = path[i].split('_')[0]
        date = path[i].split('_')[-2] + '..'
        time = path[i].split('_')[-1]
        with open(file) as f:
            lines = f.readlines()
        if (len(lines) != 0) and (lines[-4:][0].split(':')[0] == 'Video@1'):
            Video1 = float(lines[-4:][0].split(':')[1].split('\n')[0])
            VideoMean1 = float(lines[-4:][1].split(':')[1].split('\n')[0])
            Video5 = float(lines[-4:][2].split(':')[1].split('\n')[0])
            VideoMean5 = float(lines[-4:][3].split(':')[1].split('\n')[0])
            Model.append(model)
            Date.append(date)
            Time.append(time)
            Video_at_1.append(Video1)
            Video_at_MeanTop1.append(VideoMean1)
            Video_at_5.append(Video5)
            Video_at_MeanTop5.append(VideoMean5)
            eval.append(main)
            
df = pd.DataFrame({"eval" : eval, 
            "Model" : Model, 
            "Date" : Date,
            "Time" : Time, 
            "Video@1": Video_at_1,
            "Video@MeanTop1":Video_at_MeanTop1 ,
            "Video@5":Video_at_5, 
            "Video@MeanTop5":Video_at_MeanTop5} )
df['eval'].value_counts()
```

```python
models = df['Model'].unique()
evaluation = df['eval'].unique()
cols = ["Video@1","Video@MeanTop1"]
```

```python
import numpy as np
model = []
evalution_cri = []
Video1_mean , Video1_std, Video_MeanTop1_mean, Video_MeanTop1_std= [], [], [], []
for eva in evaluation:
    for mod in models:
        new_df = df[(df['Model'] == mod) & (df['eval'] == eva)]
        model.append(mod)
        evalution_cri.append(eva)
        Video1_mean.append(np.round(np.mean(new_df["Video@1"].to_numpy()),3))
        Video1_std.append(np.round(np.std(new_df["Video@1"].to_numpy()),3))
        Video_MeanTop1_mean.append(np.round(np.mean(new_df["Video@MeanTop1"].to_numpy()),3))
        Video_MeanTop1_std.append(np.round(np.std(new_df["Video@MeanTop1"].to_numpy()),3))
 
```

```python
final_df = pd.DataFrame({'criteria':evalution_cri,
                        'model':model,
                        'Video1_mean':Video1_mean,
                        'Video1_std':Video1_std,
                        'Video_MeanTop1_mean':Video_MeanTop1_mean,
                        'Video_MeanTop1_std':Video_MeanTop1_std})
final_df
```

```python
df.to_csv("results.csv")
final_df.to_csv('mean_std.csv')
```
